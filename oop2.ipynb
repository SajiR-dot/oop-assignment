```python
# Question 12

# Explain the use of abstract base classes (ABCs) in Python and how they relate to inheritance. Provide anexample using the `abc` module.

# Abstrsct bases classes are like blueprints for the child subclasses
# There are no default abstract classes in python 
# To turn a class into abstract class we need to import the ABC module
# For a class to be abstract it needs to have abstract methods
# To turn a method to abstract method it is wrapped using the @abstractmethod decorator
# We cant instantiate an abstract class
# The abstracct method has declaration but no implementation it gets overriden by the child class methods

from abc import ABC,abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    def sound(self):
        print("I bark")
        
class Cat(Animal):
    def sound(self):
        print("I purr")
        
class Tiger(Animal):
    def sound(self):
        print("I roar")
        
        
d = Dog()
d.sound()
c = Cat()
c.sound()
t = Tiger()
t.sound()        
        
# The purpose of abstract classes is to be inherited 
# if we inherit an abstract class to the subclass we should also overide the abstract method to implement it otherwise it wont be instantiated
```

```python
# Question 20

# Create a Python class for a social media application, using composition to represent users, posts, and comments.


# Creating component classes user 
class User:
    def __init__(self,user_name,password):
        self.user_name = user_name
        self.password = password
        
    def login(self,userid,userpass):
        if userid == self.user_name and userpass == self.password:
            print("Successfully Logged in")    
        else:
            print("Username or password incorrect")
        
    def logout(self):
        print("Successfully Logged out")
        

user = User("Saji2342@gmail.com","XYZ")
        
class Posts:
    def __init__(self,username):
        self.username = username
        self.posts = {} 
        
    def add_post(self,post):
        self.posts[self.username]=post
        print(f"User Name : {self.username}, Post: {post}")
         
        
            
    def remove_post(self,username):
        del self.posts[username]
        print("Post deleted successfully")
        
    
    def post_details(self):
        print(self.posts)

        
        
        
post = Posts("saji")


class Comments:
    def __init__(self,post):
        self.post = post
        self.comments_list = {}
        
        
    def add_comment(self,username,comment):
        post_name = self.post.posts[username]
        if self.comments_list == {}:
            self.comments_list[post_name]=[comment]
        else:
            self.comments_list[post_name].append(comment)
        
        print(f"You commented on the post {post_name}")
        
        
    def remove_comment(self,username):
        self.comments_list[self.post.posts[username]] = "Comment deleted "
        
        
    def comment_details(self):
        print(self.comments_list)
    
       
comments = Comments(post)



class SocialApp:
    def __init__(self,user,post,comments):
        self.obj_user = user
        self.obj_post = post
        self.obj_comments = comments
        
    def user_login(self,userid,password):
        self.obj_user.login(userid,password)
        
    def user_logout(self):
        self.obj_user.logout()
        
        
    def new_post(self,post):
        self.obj_post.add_post(post)
        
        
    def delete_post(self,username):
        self.obj_post.remove_post(username)
        
    def display_post(self):
        self.obj_post.post_details()
        
        
    def write_comment(self,username,comment):
        self.obj_comments.add_comment(username,comment)
        
    def delete_comment(self,username):
        self.obj_comments.remove_comment(username)
        
    def display_comments(self):
        self.obj_comments.comment_details()
        
        
user1 = SocialApp(user,post,comments)


user1.user_login("saji","ytyty")   
user1.user_logout()

user1.new_post("I will win")


user1.delete_post("saji")
user1.display_post()

user1.new_post("I will win")
user1.write_comment("saji","wow")
user1.display_comments()
        
```
